```python
#1. Data Context
"""  
   What is the source of the data, and how was it collected?

The source of the data is the U.S. Energy Information Administration. ^1 The data was 
collected from Electric Balancing Authorities (BA's) that operate the grid. These six 
BA's play a functional role and are mainly responsible for balancing electricity 
supply, demand, and interchange on their electric systems in real time. These six 
regional entities are the Western Electricity Coordinating Council (WECC), Midwest
Reliability Organization (MRO), Texas Reliability Entity (Texas RE), Reliability First 
(RF), Southeastern Electric Reliability Corporation (SERC) and Northeastern Power 
Coordinating Council (NPCC). Please see below for a picture of the distribution of 
the power companies.

The data was collected in different ways depending on the column of data:
1. Demand: Derived by taking the total metered net electricity generation within its
electric system and subtracting the total metered net electricity interchange occuring 
between the BA and its neighboring BAs
2. Demand by subregion: For select BAs, demand by subregion data provide demand at a 
further level of geographic disaggregation (for example, load zones, weather zones, 
operating company areas, local BAs, etc.) within a BA's electric system.
3. Demand forecast: This varies by region
4. Net generation: Derived from the metered outpout of electric generating units in a 
BA's electric system. Generators on the distribution system are typically not included.
5. Net generation by energy source: BA's generally report generation from dual-fuel and 
multiple fuel generators under the actual energy source used, if known, and under the 
generator's primary energy source, if not known.
6. Total interchange: Net metered tie line flow from one BA to another directly 
connnected BA.
7. Interchange with directly interconnected BAs:  Total net interchange is the net 
sum of all interchange occurring between a BA and its directly interconnected 
neighboring BAs.
8. Estimated CO2 emissions: Based on estimates on the hourly net generation BAs report 
to U.S. Energy Information Administration (EIA).
9. Estimated CO2 emissions by fuel type (beginning July 1, 2018): EIA calculate these 
factors for natural gas, coal, and petroleum using reported generation and our estimated 
emissions for generators operated by electric utilities and independent power producers 
within each BA fleet of generators consuming those fuels. CO2 emissions from fuels other 
than natural gas, coal, and petroleum use the same emissions factor for the entire United 
States. EIA exclude data from combined-heat-and-power facilities.
10. Estimated CO2 emissions for electricity imports or exports (beginning July 1, 2018): EIA 
estimate CO2 emissions imported from Canada and Mexico by applying the average U.S. 
emissions rate per megawatthour of generation to the megawatthours of imports.

For more information about the information that was collected or how, please visit:
https://www.eia.gov/electricity/gridmonitor/about
"""

img = mpimg.imread("Power_Companies_pic.png")   # load image
plt.imshow(img)                   # show image
plt.axis('off')                   # hide axes
plt.show()

"""
   Are there any known biases or limitations in the data?
Yes there are both in the data:
Biases include:
1. There are several generation BAs that do not directly serve retail customers. Therefore 
they do not report demand or demand forecasts
2. City of Homestead (HST) has a small number of local generators that do not always produce 
electricity, so it will not always have net generation to report.
3. Dynamic Scheduling and Pseduo ties can introduce some level of bias onto demand, demand 
forecast, net generation and interchange

Limitations
1. Reporting issues. There are four various reporting issues that could affect data quality.
This includes 
 - Limited metering data from voltage levels that vary by BAs
 - Revisions as a result of metering or telemetry issues from the immediate data read
 - Inability to read fuel consumption data on a real-time basis
 - Accounting discrepancies for transfers, pseudo-tied, and dynamically scheduled resources
2. Imputation factors. BAs occasionally report anomalous data values such as blank, zero, 
negative, and unreasonably high or low values. We perform a basic imputation process for 
several data elements including Demand, Net generation, and Total interchange
3. BA systems record the output of mostbut not allof their generators. The sum of 
these metered values for a system is called net generation.
4. This data set only accounts for data across the 48 states and does not account for
Alaska and Hawaii.


 Key Assumptions
1. the flows between areas are in balance. Due to reporting errors or discrepancies in 
 the volume of transfers between BAs, the overall flow of energy may not balance. This 
 will lead to imbalanced emissions totals.
2. Demand forecasts are typically based on prior data including daily, weekly, and 
seasonal demand patterns plus expected weatehr conditions

   How does the data relate to the problem or question you are trying to solve?
The data helps us to understand the weak points, as identified by the difference 
between expected Net generation and Demand. That will tell us which regions are 
struggling the most to provide power or forecast well. Regardless of what the reason is,
understanding that will tell us the weakest regions. Since none of the forecasts are 
based on upcoming technology requirements, I would expect these forecasts to either 
change or be very wrong.

   Important things to know:
1. Electric systems in North America are designed and operated to maintain an alternating 
current wave frequency of 60 cycles per second (hertz). Frequency is the primary measure 
that system operators use to assess the real-time state of the interconnection.
2. Deviations from 60 hertz (Hz) indicate instantaneous mismatches between electricity 
supply and demand. When frequency is higher than 60 Hz, the system has more supply than 
demand. When frequency is lower than 60 Hz, the system has more demand than available 
supply. Deviations between electricity supply and demand outside a narrow frequency band 
can result in electric system failures.
"""


"""
Sources:
^1: https://www.eia.gov/electricity/gridmonitor/about

^2: https://www.nerc.com/AboutNERC/keyplayers/Pages/default.aspx

"""
```
